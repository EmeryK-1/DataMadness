```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
from pandas.plotting import scatter_matrix
import math
df = pd.read_csv('heart.csv').set_index('HeartDisease').reset_index().rename(columns={'FastingBS':'FastingBS_120+'})
```

# CATEGORICAL FEATURES HISTOGRAM

```python
# Categorical columns (non-numeric OR only 2 values)
ohe_cols = (df.dtypes=='object')
ohe_cols = ohe_cols[ohe_cols].index.tolist()
bin_cols = (df.nunique()==2)
bin_cols = bin_cols[bin_cols].index.tolist()

size = math.ceil(math.sqrt(len(set(ohe_cols+bin_cols))))
fig, axs = plt.subplots(1, len(set(ohe_cols+bin_cols)), figsize=(20, 7))
i = 0
for col in set(ohe_cols+bin_cols):
    t = df.replace({0: 'N', 1:'Y'})
    sns.histplot(data=t, x=col, ax=axs[i])
    i+=1

plt.subplots_adjust(hspace = 0.5, wspace=0.5)
plt.show()
```

# CONTINUOUS FEATURES BOXPLOT

```python
non_bin =df.drop(columns=ohe_cols+bin_cols)
non_bin['HeartDisease'] = df['HeartDisease']
selection = non_bin.columns
fig, axs = plt.subplots(1,len(selection)-1, figsize=(13, 9))
i = 0
for col in selection:
    if col == 'HeartDisease':
        continue
    sns.boxplot(x='HeartDisease', y=col, data=non_bin, ax=axs[i])
    i+=1

plt.subplots_adjust(wspace=0.6)
plt.show()
```

# PREPARE DATASET: OHE and Binary to (0, 1)
- Sex (M or F) and ExerciseAngine (N or Y) are binary -> 0, 1
- ChestPainType, ST_Slope and RestingECG are categorical -> OneHotEncoding

```python
prepared = df.copy()
prepared[['Sex', 'ExerciseAngina']] = prepared[['Sex', 'ExerciseAngina']].replace({'M':0, 'F':1, 'N' : 0, 'Y':1})
prepared = pd.get_dummies(prepared, columns=['ChestPainType','ST_Slope', 'RestingECG'])
```
