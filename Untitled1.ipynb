```python
#necessary imports (edit as you see fit)
import numpy as np

import pandas as pd
from pandas.api.types import CategoricalDtype

%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
import math


from sklearn import linear_model as lm
from sklearn.model_selection import train_test_split
from sklearn.model_selection import KFold

# Plot settings
plt.rcParams['figure.figsize'] = (12, 9)
plt.rcParams['font.size'] = 12
```

```python
df = pd.read_csv("heart.csv")
```

```python
df.info()
```

```python
ohe_cols = (df.dtypes=='object')
ohe_cols = ohe_cols[ohe_cols].index.tolist()
bin_cols = (df.nunique()==2)
bin_cols = bin_cols[bin_cols].index.tolist()

size = math.ceil(math.sqrt(len(set(ohe_cols+bin_cols))))
fig, axs = plt.subplots(1, len(set(ohe_cols+bin_cols)), figsize=(20, 7))
i = 0
for col in set(ohe_cols+bin_cols):
    t = df.replace({0: 'N', 1:'Y'})
    sns.histplot(data=t, x=col, ax=axs[i])
    i+=1

plt.subplots_adjust(hspace = 0.5, wspace=0.5)
plt.show()
```

```python
non_bin =df.drop(columns=ohe_cols+bin_cols)
non_bin['HeartDisease'] = df['HeartDisease']
selection = non_bin.columns
fig, axs = plt.subplots(1,len(selection)-1, figsize=(13, 9))
i = 0
for col in selection:
    if col == 'HeartDisease':
        continue
    sns.boxplot(x='HeartDisease', y=col, data=non_bin, ax=axs[i])
    i+=1

plt.subplots_adjust(wspace=0.6)
plt.show()
```

```python
cnt = non_bin.drop('HeartDisease', 1)
fig, axs = plt.subplots(len(cnt.columns), figsize=(5, 25))
for i, col in enumerate(cnt.columns):
    cnt[col].hist(ax=axs[i])
    
plt.subplots_adjust(wspace=0.6)
plt.show()
```

```python
cnt1 = cnt.drop('Oldpeak',1)
for i, col in enumerate(cnt1.columns):
    sns.distplot(cnt1[col])
    
plt.legend(labels = ["Age", "RestingBP",'Cholesterol','MaxHR'])    
    
```

```python
cate = df.select_dtypes(include=['object']).copy()
print(cate.isnull().sum())
cate
```

```python
print(cate['Sex'].value_counts())
print(cate['ChestPainType'].value_counts())
print(cate['RestingECG'].value_counts())
print(cate['ExerciseAngina'].value_counts())
print(cate['ST_Slope'].value_counts())
```

```python
new_s = {'Sex': {'M': 0, 'F': 1}}
new_s
labels = cate['Sex'].astype('category').cat.categories.tolist()
replace_map_s = {'Sex' : {k: v for k,v in zip(labels,list(range(1,len(labels)+1)))}}

cat_replace = cate.copy()
cat_replace.replace(replace_map_s, inplace=True)

print(cat_replace.head())
```
